```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pyperclip
from selenium.webdriver.support.ui import Select
from datetime import datetime, timedelta
import calendar

###############################################################################
############################id,pw설정##########################################
ID = ''
PW = ''
################################################################################
################################################################################






options = Options()
#최대화면으로 설정
options.add_argument('--start-maximized')
#꺼짐 방지
options.add_experimental_option("detach",True)
service = Service(ChromeDriverManager().install())
#크롬드라이버 설정
driver = webdriver.Chrome(service = service, options=options)
#URL 설정
url1 = "https://www.airportal.go.kr/common/login/login01.jsp"
#URL 접속
driver.get(url1)
time.sleep(2)
#유저 아이디 찾아 아이디 전송(차단 먹을 수 있고 이는 COPY_KEY사용하여 해결 가능)
driver.find_element(By.NAME, 'df_userid').send_keys(ID)
time.sleep(2)
#유저 비번 찾아 비밀번호 전송(차단 먹을 수 있고 이는 COPY_KEY사용하여 해결 가능)
driver.find_element(By.NAME, 'df_passwd').send_keys(PW)
time.sleep(2)
#로그인버튼 찾고 클릭
login_button = driver.find_element(By.CSS_SELECTOR, "input[type='image'][alt='로그인 버튼']")
login_button.click()
time.sleep(2)
#실시간 출도착 정보 들어가기
driver.find_element(By.LINK_TEXT, '실시간출도착정보').click()
time.sleep(2)
#엑셀 다운로드 들어가기
driver.find_element(By.ID,'excel_div').click()
time.sleep(5)
##중요! 접속할 때 창이 바뀌므로 셀레니움이 바라보는 창 바꾸기
main_handle= driver.window_handles
driver.switch_to.window(driver.window_handles[1])
time.sleep(1)
#공항 찾고 클릭
driver.find_element(By.NAME,'airport').click()
time.sleep(3)
inputs = [Keys.ARROW_DOWN]*7
#엔터 클릭
driver.find_element(By.NAME,'airport').send_keys(*inputs,Keys.ENTER)
#시작 날짜 설정
start_date = datetime(2010, 1, 1)
#종료 날짜 설정
end_date = datetime(2023, 6, 1)
current_date = start_date
time.sleep(2)


driver.find_elements(By.CLASS_NAME,'iCheck-helper')[1].click()
while current_date <= end_date:
    _, last_day = calendar.monthrange(current_date.year, current_date.month)
    start_str = current_date.strftime("%Y%m%d")
    end_str = current_date.replace(day=last_day).strftime("%Y%m%d")
    #사이트가 2번 클릭해야 바뀌므로 2번 클릭
    driver.find_element(By.NAME,'sDate').click()
 
    driver.find_element(By.NAME,'sDate').clear()
  
    driver.find_element(By.NAME,'sDate').send_keys('20230501',Keys.ENTER)

    driver.find_element(By.NAME,'sDate').click()

    inputs = [Keys.ARROW_RIGHT]*10 + [Keys.BACK_SPACE]*10
    driver.find_element(By.NAME,'sDate').send_keys(*inputs)

    driver.find_element(By.NAME,'sDate').send_keys(start_str,Keys.ENTER)

    driver.find_element(By.NAME,'eDate').click()
    #사이트가 2번 클릭해야 바뀌므로 2번 클릭
    driver.find_element(By.NAME,'eDate').send_keys('20230601',Keys.ENTER)

    driver.find_element(By.NAME,'eDate').click()

    driver.find_element(By.NAME,'eDate').send_keys(*inputs)

    driver.find_element(By.NAME,'eDate').send_keys(end_str,Keys.ENTER)
    #검색 버튼 누르기
    driver.find_element(By.CLASS_NAME,'mainSearchBtn').click()
    time.sleep(8)
    #엑셀 다운 버튼 누르기
    driver.find_element(By.CLASS_NAME,'mainExcellBtn').click()
    time.sleep(10)
    if current_date.month == 12:
        current_date = current_date.replace(year=current_date.year + 1, month=1, day=1)
    else:
        current_date = current_date.replace(month=current_date.month + 1, day=1)


```
